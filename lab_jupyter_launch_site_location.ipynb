```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # approximate radius of earth in km

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

proximity_data = {
    'Ocean': {'Lat': 28.56, 'Long': -80.57},
    'City': {'Lat': 28.61, 'Long': -80.81},
}

for index, launch_site in launch_sites_df.iterrows():
    launch_lat = launch_site['Lat']
    launch_lon = launch_site['Long']
    for proximity_name, proximity_coord in proximity_data.items():
        proximity_lat = proximity_coord['Lat']
        proximity_lon = proximity_coord['Long']
        distance = calculate_distance(launch_lat, launch_lon, proximity_lat, proximity_lon)
        print(f"{launch_site['Launch Site']} to {proximity_name}: {distance:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
